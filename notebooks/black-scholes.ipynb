# Black scholes equation

The **Black-Scholes** equation is used to price options and other derivative securities (e.g. currency options, futures and commodity options,  interest ratte derivates, ESOs, exotic options, variance and volatility derivatives). It was developed by Fischer Black, Myron Scholes, and later refined by Robert Merton in the early 1970s, leading to the famous Black-Scholes-Merton Model.

This is a partial differential equation that describes how the price of an evolves over time. Given by:
$$
\frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + rS \frac{\partial V}{\partial S} - rV = 0
$$

where:
* $V$ = price option
* $S$ = price of the underlying asset (e.g. stock price)
* $t$ = time
* $\sigma$ = volatility of the underlying asset
* $r$ = risk-free interest rate

### Formula 

For **European call options** the formula is: 
$$
C = S_0 N(d_1) - Ke^{rt}N(d_2)
$$

where:
* $C$ = price of the European call option
* $S_0$​ = current stock price
* $K$ = strike price
* $T$ = time to expiration
* $r$ = risk-free interest rate
* $\sigma$ = volatility
* $N(d)$ = cumulative normal distribution function
* $d_1$ = $\frac{\ln(S_0 / K) + (r+ \sigma^2 /2)T }{\sigma \sqrt{T}}$
* $d_2$ = $d_1- \sigma \sqrt{T}$
​

### Assumptions of the Model 

1. The stock price follows a geometric Brownian motion with constant drift (i.e. steady rate of growth) and volatility.
2. The market is frictionless, meaning no transaction costs or taxes.
3. No arbitrage opportunities exist.
4. The option can only be exercised at expiration (European-style options).
5. The risk-free interest rate is constant over time.
6. The stock does not pay dividends (though adjustments exist for dividend-paying stocks).

### Limitations

* It assumes constant volatility, the real market doen's really work that well. 
* It does not account for liquidity risks or market impact.
* It does not work well for American options, which can be exercised before expiration.

```python
# the strike price K
S = 50
K = list(range(101,200))
T = 10
r = 0.05
sigma = 0.2

call_prices = [black_scholes_equation(S, strike,T,r,sigma,'c') for strike in K]

visualizer_bs_call(K, call_prices, 'strike prices')
```

```python
# the time to maturity T
S = 100
K = 110
T = list(range(1,100))
r = 0.05
sigma = 0.2

call_prices = [black_scholes_equation(S, K,t,r,sigma,'c') for t in T]

visualizer_bs_call(T, call_prices, 'time to maturity')
```

```python
# the risk free rate r
S = 100
K = 110
T = 10
r = list(np.arange(0.01,0.5,0.01))
sigma = 0.2

call_prices = [black_scholes_equation(S, K,T,rate,sigma,'c') for rate in r]

visualizer_bs_call(r, call_prices, 'risk free rate')
```

```python
# the volatility
S = 100
K = 110
T = 10
r = 0.05
sigma = list(np.arange(0.1,1,0.01))

call_prices = [black_scholes_equation(S, K,T,r,v,'c') for v in sigma]

visualizer_bs_call(sigma, call_prices, 'volatility')
```
