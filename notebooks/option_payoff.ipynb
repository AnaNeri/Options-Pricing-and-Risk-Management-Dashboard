# Option payoff 

An option payoff refers to the profit or loss an option holder experiences at expiration, depending on the underlying asset's price. 

### Option payoff ignoring costs

When ignoring costs (such as the option premium and transaction fees), the payoff is based solely on the intrinsic value of the option at expiration.

Call:
$$ Payoff = max (S - K, 0)$$

Put:
$$ Payoff = max( K-S, 0)$$

e.g. If one owns a call with $K=50$ and the stock ends at $S=60$, the payoff is $\$10$.

### Option payoff with cost

When including costs (like the option premium $P$), the profit/loss equation becomes:

Call:
$$ Payoff = max (S - K, 0) -P $$

Put:
$$ Payoff = max( K-S, 0) -P $$

### Protective Puts (Hedging downside risk)

A protective put is when an investor buys a put option while also owning the stock. This strategy limits downside losses while keeping upside potential.

Formula:
$$Payoff = max(K-S,0) + S-P$$

* If stock price drops, the put protects by increasing in value.
* If stock price rises, the put expires worthless, but there is profit from the stock.

### Fiduciaty Call 

A fiduciary call is a combination of buying a call option and investing the present value of the strike price in a risk-free bond. It replicates a protective put strategy.
